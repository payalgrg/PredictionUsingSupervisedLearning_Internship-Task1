## <span style="font-family:Times New Roman">What will be the predicted score if a student studies for 9.25 hours a day?</span>

### <span style="font-family:Times New Roman">Importing necessary libraries</span>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

### <span style="font-family:Times New Roman">Reading the Data </span>

```python
data = pd.read_csv('http://bit.ly/w-data')
data.head(5)
```

```python
# Check if there any null value in the Dataset
data.isnull == True
```

### <span style="font-family:Times New Roman">There is no null value in the Dataset so, we can now visualize our Data</span>

```python
sns.set_style('white')
sns.scatterplot(y= data['Scores'], x= data['Hours'])
plt.title('Percentage vs Hours',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

**<span style="font-family:Times New Roman">From the above scatter plot there looks to be correlation between the 'Marks Percentage' and 'Hours Studied', Lets plot a regression line to confirm the correlation.</span>**

```python
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
print(data.corr())
```

**<span style="font-family:Times New Roman">It is confirmed that the variables are positively correlated.</span>**

### <span style="font-family:Times New Roman">Training the Model</span>
### <span style="font-family:Times New Roman">1) Splitting the Data</span>

```python
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

### <span style="font-family:Times New Roman">2) Fitting the Data into the model</span>

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("---------Model Trained---------")
```

### <span style="font-family:Times New Roman">Predicting the Percentage of Marks</span> 

```python
pred_y = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

### <span style="font-family:Times New Roman">Comparing the Predicted Marks with the Actual Marks</span> 

```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

### <span style="font-family:Times New Roman">Visually Comparing the Predicted Marks with the Actual Marks</span> 

```python
plt.scatter(x=val_X, y=val_y, color='blue')
plt.plot(val_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

### <span style="font-family:Times New Roman">Evaluating the Model</span> 

```python
# Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

*** <span style="font-family:Times New Roman">Small value of Mean absolute error states that the chances of error or wrong forecasting through the model are very less.</span> ***

### <span style="font-family:Times New Roman">What will be the predicted score if a student studies for 9.25 hrs/ day?</span> 

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

## <span style="font-family:Times New Roman">Therefore, according to the regression model if a student studies for 9.25 hours a day he/she is likely to score 93.89 marks.</span> ##
